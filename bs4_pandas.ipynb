```python
import requests
from bs4 import BeautifulSoup
import re

url = 'https://feeds.buzzsprout.com/1974862.rss'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

items = soup.find_all('item')

result_list = []

for item in items:

    description = item.find('description').get_text(strip=True)

    # 去除 HTML 標籤
    description = re.sub('<.*?>', '', description)


    matches = re.findall('===(.*?)===', description, re.DOTALL)

    for match in matches:
        match = match.strip()
        if '專屬連結' in match or '專屬優惠' in match:
            result_list.append(match)

        if len(result_list) >= 8:
            break

    if len(result_list) >= 8:
        break

for result in result_list:
    print(result)
    print()
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://feeds.buzzsprout.com/1974862.rss'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

keyword = input("請輸入關鍵字：")

items = soup.find_all('item')
count = 0  # 用於計數已顯示的結果數量

for item in items:
    title = item.find('title').text
    description = item.find('description').get_text(strip=True)

    if keyword.lower() in title.lower() or keyword.lower() in description.lower():
        print('標題:', title)
        print('簡述:', description)
        print('---')
        count += 1

        if count >= 10:  # 當顯示數量達到 10 筆時停止顯示
            break
```

# 松果購物

```python
import requests
import json
from prettytable import PrettyTable
import IPython.display as dp

table = PrettyTable(["名稱", "價格"], encoding="utf-8")
table.align["名稱"] = "l"
table.align["價格"] = "l"

word = input("關鍵字：")
page = 1

while True:
    try:
        response = requests.get("https://xapi.pcone.com.tw/search/commodities",headers={
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.190 Safari/537.36",
            "Cookie": "PHPSESSID=qlemnrneki3bc204e83thcej4h"
        },params={
            'site':'pcone',
            'page':page,
            'q':word,
            'sorting':'default',
            'device':'web',
            'aid':'null'
        })

        data = json.loads(response.text)

        table.clear_rows()

        for prod in data["data"]["commodities"]:
            table.add_row([prod["name"], prod["price"]])

        print(table)

        page = input("前往頁碼：")

        dp.clear_output(True)

    except KeyboardInterrupt:
        break
```

# PTT

```python
from bs4 import BeautifulSoup
import requests
url = "https://www.ptt.cc/bbs/movie/M.1685861654.A.409.html"
response = requests.get(url)
html = BeautifulSoup(response.text)


content = html.find("div" , id = "main-content")

metas = content.find_all("span",class_="article-meta-value")
print("ID",metas[0].text)
print("看板",metas[1].text)
print("標題",metas[2].text)
print("時間",metas[3].text)

ms = content.find_all("div",class_="article-metaline")
for m in ms:
    m.extract()

ms = content.find_all("div",class_="article-metaline-right")
for m in ms:
    m.extract()

pushes = content.find_all("div",class_ = "push")
score = 0
for p in pushes:
    tag = p.find("span",class_="push-tag").text
    if "推" in tag:
        score = score +1
    elif "噓" in tag:
        score = score -1
    p.extract()
print("分數：",score)
print("本文",content.text)
```
